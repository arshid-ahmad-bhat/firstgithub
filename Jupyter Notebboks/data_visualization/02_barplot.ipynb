```python
#import liberairies
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 
#load dataset
phool = sns.load_dataset("iris")
phool
```

```python
# draw a barplot
sns. barplot(x="species", y="sepal_length", data = phool)
plt.show()
```

```python
sns.barplot(x="species", y= "petal_length", data = phool)
plt.show()
```

```python
#setting order of varibles and setiig color
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "sex", y="alone", hue = "who", data = kashti, order=["female", "male"], palette = 'dark:none')
plt.show()
```

```python
# removing error bars by setting errorbar = None
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "sex", y="alone", hue = "who", data = kashti, order=["female", "male"], palette= "dark:red", errorbar= None)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "sex", y="alone", hue = "who", data = kashti, order=["female", "male"], palette= "dark:red", errorbar= None, estimator = mean )
plt.show()
```

```python
#color saturation|intensity
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "sex", y="alone", hue = "who", data = kashti, order=["female", "male"], palette= "dark:red", errorbar= None, saturation= 0.1)
plt.show()
```

```python
#color saturation|intensity
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "sex", y="alone", hue = "who", data = kashti, order=["female", "male"], palette= "dark:red", errorbar= None, saturation= 3)
plt.show()
```

```python
# horizantal plot(by ploting catagorical varibale on y axis and numerical on x axis)
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "fare", y="class", hue = "sex", data = kashti, palette= "dark:red", saturation= 1)
plt.show()
```

```python
# Setting line width(Edges) , colors of edges, bars and faces of plot
import seaborn as sns 
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
sns.barplot(x= "class", y="fare", data = kashti,
           linewidth = 3, facecolor=(0.1,0.3,0.3,0.5),
            errcolor="0.5", edgecolor="0")
plt.show()
```
