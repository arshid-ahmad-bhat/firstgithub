# adding titles

```python
#import liberaies 
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
# draw a line plot 
sns.lineplot(x="sepal_length", y = "sepal_width", data=phool)
plt.title("PHOULOUN KA PLOT")
plt.show()
```

# adding limits

```python
#import liberaies 
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
# draw a line plot 
sns.lineplot(x="sepal_length", y = "sepal_width", data=phool)
plt.title("PHOULOUN KA PLOT")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# Setting Styles

```python
#import liberaies 
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool
# draw a line plot 
sns.lineplot(x="sepal_length", y = "sepal_width", data=phool)
plt.title("PHOULOUN KA PLOT")
sns.set_style("dark")
plt.show()
```

# Size of FIGURe

```python
# #import liberaies 
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

#change figure
plt.figure(figsize=(3,3))
# draw a line plot 
sns.lineplot(x="sepal_length", y = "sepal_width", data=phool)
plt.title("PHOULOUN KA PLOT")
sns.set_style("dark")
plt.show()
```
