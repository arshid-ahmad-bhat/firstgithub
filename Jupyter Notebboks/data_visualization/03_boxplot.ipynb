### Managing individual colors for each hue catagory
- 1. Defining Custom palette
- 2. Using Seaborn’s Built-in Palettes
- 3. Using a List of Colors
- 4. Using sns.color_palette()

```python
# Definig Custom palette 
import seaborn as sns

# used to set style of background of plot
sns.set(style='whitegrid')

#loading the dataset
tip = sns.load_dataset("tips")

# Define custom colors for each hue category
custom_palette = {
    "Male": "blue",
    "Female": "red"
}

# Create boxplot with custom colors
sns.boxplot(data=tip, x="day", y="total_bill", hue="sex", palette=custom_palette)

```

```python
# Using list of colors
import seaborn as sns

# used to set style of background of plot
sns.set(style='whitegrid')

#loading the dataset
tips = sns.load_dataset("tips")

sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette=["#FF5733", "#33FF57"])
```

```python
# using seaborn predefined palette
import seaborn as sns

# used to set style of background of plot
sns.set(style='whitegrid')

#loading the dataset
tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="coolwarm")
```

### Customising plots

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style= "whitegrid")
kashti = sns.load_dataset("titanic")
kashti.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set(style= "whitegrid")
kashti = sns.load_dataset("titanic")
sns.boxplot(data=kashti , x= "survived" , y= "age", palette= ["red","green"])
```

```python
#showing mean explicitly with symbols
sns.boxplot(x='survived' , y ='age', 
             showmeans= True ,
           meanprops = {"marker" : "*",
                       "markersize" :"12",
                       "markeredgecolor" : "red"},
                        data =  kashti)
```

```python
#setting labels
sns.boxplot(x='survived' , y ='age', 
             showmeans= True ,
           meanprops = {"marker" : "*",
                       "markersize" :"12",
                       "markeredgecolor" : "red"},
                        data =  kashti)
plt.xlabel("How many survived", size = 10, weight = 'bold'),
plt.ylabel("Age(years)", size=12, weight = 'bold')
plt.title("Box Plot Of kitnay DOOB gayay or kitnay bach gayay", size = 16, weight = 'bold')
plt.show()
```
