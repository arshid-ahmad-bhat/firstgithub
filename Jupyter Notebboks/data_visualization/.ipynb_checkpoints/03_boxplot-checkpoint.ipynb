```python
#We draw numerical variableson y-axis and catagorical varaibles on x-axis or in hue , numnerical varible will never be drawn in hue.
import seaborn
#canvas (Baloon board)
seaborn.set(style= 'whitegrid')
kashti = seaborn.load_dataset("titanic")
kashti
seaborn.boxplot(x="class", y = "fare", data = kashti)
```

```python
import seaborn as sns
sns.set(style="whitegrid")
tip = sns.load_dataset('tips')
tip
sns.boxplot(x="day", y="tip", data = tip, saturation = 0.3)
```
