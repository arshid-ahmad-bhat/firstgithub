## Steps involved in data visualizationn

```python
# Step 1 - import liberaies
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2 - set a theme
sns.set_theme(style = "ticks", color_codes = True)
```

```python
# Step 3- import data set inbuilt|own
titanic =sns.load_dataset("titanic")

# Step 4 - plot basic graph with 1 variable (count)
p = sns.countplot(x="sex", data = titanic)
plt.show()
```

```python
# Step-5 plot basic graph with two variables (count plot)
p = sns.countplot(x="sex", data = titanic , hue = "class")
plt.show()
```

```python
# Step-5 plot basic graph with two variables (count plot) with title
p = sns.countplot(x="sex", data = titanic , hue = "class")
p.set_title("Arshid ka cout plot for titanic")
plt.show()
```

```python
# #Then why assign p = sns.countplot() at all?
# By assigning the count plot to p, we store the plot object. This allows us to modify it later (like adding a title, labels, or other customizations).
# If we didn't assign it to p, we wouldn't be able to modify the plot afterward.

sns.countplot(x="sex", data=titanic)
plt.show()

sns.countplot(x="sex", data=titanic, hue="class")
plt.show()
```
